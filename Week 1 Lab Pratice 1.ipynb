Model representation 

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('./deeplearning.mplstyle')
```

```python
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f"x_train = {x_train}, y_train = {y_train}")
```

```python
m = x_train.shape[0]

print(f"Number of training examples: {m}")
```

Training examples

```python
for i in range(m):
    x_i = x_train[i]
    y_i = y_train[i]
    print(f"When i = {i}, x^({i}) = {x_i} and y^({i}) = {y_i}")
```

```python
plt.scatter(x_train, y_train, marker = "o", c="g")
plt.title("Housing Prices")
plt.ylabel("Size (1000 sqft)")
plt.xlabel("Price")
plt.show()
```

```python
w = 100
b = 100
print(f"w : {w}, b: {b}")
```

```python
def compute_model_output(x, w, b):
    
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
        
    return f_wb       
```

```python
prediction = compute_model_output(x_train, w, b)

# Model prdeictions
plt.plot(x_train, prediction, marker = "x", c="r", label="My Predictions")

# Actual value
plt.scatter(x_train, y_train , marker="o", c="g", label="Actual value")

plt.title("Housing Prices")

plt.xlabel("Size")

plt.ylabel("Price")

plt.legend()

plt.show()
```
